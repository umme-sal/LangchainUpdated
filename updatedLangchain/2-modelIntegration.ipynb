Streaming and Batch

```python
from langchain.chat_models import init_chat_model

model=init_chat_model("groq:llama-3.3-70b-versatile")
# response=model.invoke("distance between earth and sun")
# response.content
for chunk in model.stream("Write me a 200 words of paragraph on AI"):
    print(chunk.text,end="",flush=True)
```

```python
model.batch([
    "Who is the PM of USA",
    "What is quantum computing"
],
    config={
        'max_concurrency':5, #limit to 5 || calls
    }  
)
```
